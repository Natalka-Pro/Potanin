```python
FOLDER_CLEAR = "КККМ_clear/КККМ/"
FOLDER = "КККМ"

def max_common_part(name1, name2):
    if name1 == name2:
        return float("inf")
    else:

        for i in range(min(len(name1), len(name2))):
            if name1[i] != name2[i]:
                break
        return i
        
list_images_clear = sorted(os.listdir(FOLDER_CLEAR))
list_images = sorted(os.listdir(FOLDER))
print(len(list_images_clear), len(list_images))

name2clear = {}
possible_names = set(list_images_clear)

for name in list_images:

    sim = 0
    near_clear_name = None
    for clear_name in possible_names:

        cur_sim = max_common_part(name, clear_name)
        if cur_sim > sim:
            sim = cur_sim
            near_clear_name = clear_name

    name2clear[name] = (near_clear_name, sim)
    possible_names.discard(near_clear_name) # error not raised

print(possible_names)
# КККМ ОФ 7928:817 Монякова.jpeg
# КККМ ОФ 7928:817 Муромов.jpeg
```

```python
for name, (clear_name, sim) in sorted(name2clear.items()):
    if sim != float("inf"):
        print(f"# {clear_name} -> {name}")
```
